## 이화여대 전체 공지사항 스크래핑

**스크래핑**

```python
# User-Agent 헤더 설정
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/98.0.4758.102"}

# 제목 저장 리스트 생성
title_list = []

# 첫 페이지와 그 이후 페이지를 구분하여 스크래핑
def web_scraping(start_page, end_page):
    for page_number in range(start_page, end_page + 1):
        print(f'👉 페이지 번호: {page_number}')

        # 오프셋 계산
        offset = (page_number - 1) * 10
        url = f'http://www.ewha.ac.kr/ewha/news/notice.do?mode=list&&articleLimit=10&article.offset={offset}'

        # 웹 페이지 요청 및 HTML 소스 가져오기
        req = requests.get(url, headers=headers)
        html_content = req.text

        # BeautifulSoup 객체 생성
        soup = BeautifulSoup(html_content, 'html.parser')

        #'제목'만 가져오기
        titles = soup.find_all(attrs={'class': 'b-title-box'})

        # 첫 페이지에서는 모든 공지사항을 포함
        if page_number == 1:
            for title in titles:
                title_text = ' '.join(title.stripped_strings)
                print(title_text)
                title_list.append(title_text)
        # 이후 페이지에서는 고정 공지사항을 제외하고 가져오기
        else:
            for title in titles[8:]:
                title_text = ' '.join(title.stripped_strings)
                print(title_text)
                title_list.append(title_text)
```

```python
# 1페이지부터 32페이지까지 스크래핑
web_scraping(1, 32)

# 결과 출력
print(title_list)
```

```python
# User-Agent 헤더 설정
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/98.0.4758.102"}

# 제목 저장 리스트 생성
title_list = []

# 첫 페이지와 그 이후 페이지를 구분하여 스크래핑
def no_page_web_scraping(start_page, end_page):
    for page_number in range(start_page, end_page + 1):
        # 오프셋 계산
        offset = (page_number - 1) * 10
        url = f'http://www.ewha.ac.kr/ewha/news/notice.do?mode=list&&articleLimit=10&article.offset={offset}'

        # 웹 페이지 요청 및 HTML 소스 가져오기
        req = requests.get(url, headers=headers)
        html_content = req.text

        # BeautifulSoup 객체 생성
        soup = BeautifulSoup(html_content, 'html.parser')

        #'제목'만 가져오기
        titles = soup.find_all(attrs={'class': 'b-title-box'})

        # 첫 페이지에서는 모든 공지사항을 포함
        if page_number == 1:
            for title in titles:
                title_text = ' '.join(title.stripped_strings)
                title_list.append(title_text)
        # 이후 페이지에서는 고정 공지사항을 제외하고 가져오기
        else:
            for title in titles[8:]:
                title_text = ' '.join(title.stripped_strings)
                title_list.append(title_text)
```

```python
# 1페이지부터 32페이지까지 스크래핑 수행
no_page_web_scraping(1, 32)

# 결과 출력
print(title_list)
```
